```python
#import all necessary modules from keras & python packages
#
import os
import keras
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt

```

```python
#img_path = ".\Downloads\img\cat.jpg"
img_path = r"C:\Users\Maitreyee\Downloads\image"
img_file = r"cat.jpg"
img_url=os.path.join(img_path, img_file)
print("URL:" + str(img_url))
#img_path = os.path.join(".", "Downloads", "img", "cat.jpg")
#img_path1=img_path.replace('\\\\','\\')
#print(img_path)
#print(img_path1)
img=image.load_img(img_url, target_size=(224, 224))
#print(img)
plt.imshow(img)
plt.show()
```

Learning  : 
Using direct path string might cause escape character issue (\\ double backslash in path ) while os.path.join with path & filename provided as argument will get rid of such issues.
How images store information : 
An image is collection of pixels laid out in rectangular grid


```python
#since I have only 1 image & I want a batch of images
img_array = image.img_to_array(img)
img_batch = np.expand_dims(img_array, axis=0 )
```

```python
model = tf.keras.applications.resnet50.ResNet50()
```

```python
img_preprocessed = preprocess_input(img_batch)
```

```python
prediction = model.predict(img_preprocessed)
print(decode_predictions(prediction, top=10)[0])
```
